1.     Define a class Complex that defines a complex number with attributes real and imaginary (as we did in the class). Define operators for addition, subtraction, multiplication and division (should work for ‘/’ and ‘//’) (Do with both operator overloading as well as without overloading). While printing the output, print in the form of complex number form like ( a + ib)

```python
class Complex:
  def __init__(self,real,imaginary):
    self.real=real
    self.imag=imaginary
#without operator overloading-
  def add(self,other):
    sum=Complex(self,other)
    sum.real=self.real+other.real
    sum.imag=self.imag+other.imag
    print(f"The sum is {sum.real}+i({sum.imag})")
  
  def sub(self,other):
    diff=Complex(self,other)
    diff.real=self.real-other.real
    diff.imag=self.imag-other.imag
    print(f"The difference is {diff.real}+i({diff.imag})")

  def mul(self,other):
    prod=Complex(self,other)
    prod.real= self.real*other.real - self.imag*other.imag
    prod.imag= self.real*other.imag+ self.imag*other.real
    print(f"The product is {prod.real}+i({prod.imag})")

  def div(self,other):
    quo=Complex(self,other)
    quo.real=(self.real*other.real+self.imag*other.imag)/(other.real**2+other.imag**2)
    quo.imag=(self.imag*other.real-self.real*other.imag)/(other.real**2+other.imag**2)
    print(f"The quotient is {quo.real}+i({quo.imag})")
  
  def div_float(self,other):
    quo_f=Complex(self,other)
    quo_f.real=(self.real*other.real+self.imag*other.imag)//(other.real**2+other.imag**2)
    quo_f.imag=(self.imag*other.real-self.real*other.imag)//(other.real**2+other.imag**2)
    print(f"The quotient using float division is {quo_f.real}+i({quo_f.imag})")

  #using operator overloading
  def __add__(self,other):
    sum1=Complex(self,other)
    sum1.real=self.real+other.real
    sum1.imag=self.imag+other.imag
    print(f"The sum is {sum1.real}+i({sum1.imag})")

  def __sub__(self,other):
    diff1=Complex(self,other)
    diff1.real=self.real-other.real
    diff1.imag=self.imag-other.imag
    print(f"The difference is {diff1.real}+i({diff1.imag})")

  def __mul__(self,other):
    prod1=Complex(self,other)
    prod1.real= self.real*other.real - self.imag*other.imag
    prod1.imag= self.real*other.imag+ self.imag*other.real
    print(f"The product is {prod1.real}+i({prod1.imag})")

  def __truediv__(self,other):
    quo1=Complex(self,other)
    quo1.real=(self.real*other.real+self.imag*other.imag)/(other.real**2+other.imag**2)
    quo1.imag=(self.imag*other.real-self.real*other.imag)/(other.real**2+other.imag**2)
    print(f"The quotient is {quo1.real}+i({quo1.imag})")

  def __floordiv__(self,other):
    quo1_f=Complex(self,other)
    quo1_f.real=(self.real*other.real+self.imag*other.imag)//(other.real**2+other.imag**2)
    quo1_f.imag=(self.imag*other.real-self.real*other.imag)//(other.real**2+other.imag**2)
    print(f"The quotient using float division is {quo1_f.real}+i({quo1_f.imag})")



C1=Complex(4,8)
C2=Complex(1,3)
C1.add(C2)
C1.sub(C2)
C1.mul(C2)
C1.div(C2)
C1.div_float(C2)
C1+C2
C1-C2
C1*C2
C1/C2
C1//C2
```
