# AlexNet and ImageNet
- AlexNet is a type of deep learning neural network architecture developed in 2012 by Alex Krizhevsky, Ilya Sutskever, and Geoffrey Hinton. It was one of the first neural networks to successfully use convolutional layers to recognize objects in images, and it was trained on a large dataset called ImageNet.
- The ImageNet dataset contains millions of labeled images belonging to one of 1000 categories. The categories are diverse and include objects, animals, plants, and more. The dataset is split into three sets: training, validation, and testing. The training set contains over a million images, while the validation and testing sets contain tens of thousands of images each.
- AlexNet was trained on the ImageNet dataset using a technique called supervised learning. During training, the model learns to classify images into one of the 1000 categories by adjusting its internal weights based on the error between its predictions and the correct labels.
- Overall, the AlexNet dataset refers to the ImageNet dataset used to train the AlexNet architecture, which has been instrumental in advancing the field of computer vision and deep learning.

# ShapeNet
- ShapeNet is a large-scale 3D object dataset that contains over 55 categories of objects, each represented as a collection of 3D models. The dataset was created to provide a benchmark for 3D object recognition and retrieval tasks.
- The ShapeNet dataset is widely used for tasks such as 3D object recognition, classification, and retrieval. The large number of categories and models in the dataset provide a diverse range of objects with varying shapes and structures, making it a valuable resource for training and evaluating 3D deep learning models.


# ModelNet
- ModelNet is a large-scale 3D object dataset that contains over 127,000 3D CAD models belonging to 55 object categories. The objects in the dataset are represented as triangular meshes, and each object is manually annotated with its corresponding category label.
- The objects in the dataset are labeled with their corresponding category, such as chair, table, or lamp, and the dataset is split into a training set and a testing set. The training set contains over 120,000 models, while the testing set contains over 7,000 models.
-  The ModelNet dataset has been used in various research studies to evaluate the performance of 3D deep learning models.

# CADSketchnet
- CadSketchNet is a 2D and 3D object dataset that contains hand-drawn sketches of objects in various categories, such as furniture, electronics, and vehicles. The dataset was created to provide a benchmark for sketch-based 3D object retrieval and classification tasks.
- The objects in the CadSketchNet dataset are represented as sketches created by human users using a software tool that simulates drawing with a pen on a tablet. Each sketch is accompanied by metadata, such as the object category, the number of strokes used to draw the sketch, and the number of keypoints or features detected in the sketch.
