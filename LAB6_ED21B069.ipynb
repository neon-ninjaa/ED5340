1.      Create a class called Matrix with methods to perform various matrix operations using the Numpy library. The class should take two matrices as input attributes and implement the following methods: 

a)      elementwise_addition()

b)      elementwise_subtraction()

c)      elementwise_multiplication()

d)      scalar_multiplication(scalar)

e)      determinant()

f)       matrix_multiplication()

g)      transpose()

h)      inverse()

i)       diagonal(): Print the diagonal elements of the first matrix.

j)       rank()

k)      eigen() : Calculate the eigenvalues and eigenvectors of the first matrix

Note: The eigen() method should also plot the eigenvectors using Matplotlib, which should be imported and used inside the Matrix class.

If the matrices are incompatible for a particular operation or the matrix is not square for calculating the determinant or not invertible for calculating the inverse, a ValueError should be raised

```python
import numpy as np
import matplotlib.pyplot as plt
class Matrix:
    def __init__(self,mat1,mat2):
        self.m1=mat1
        self.m2=mat2
    def elementwise_add(self):
        if self.m1.shape==self.m2.shape:
            return (np.add(self.m1,self.m2))
        else:
            print("The matrices cannot be added")
    def elementwise_sub(self):
         if self.m1.shape==self.m2.shape:
            return (np.subtract(self.m1,self.m2))
         else:
            print("The matrices cannot be subtracted")
    def elementwise_mult(self):
        if self.m1.shape==self.m2.shape:
            return (np.multiply(self.m1,self.m2))
        else:
            print("The matrices cannot be multiplied elementwise")
    def scalar_mult(self,num):
        self.n=num
        return (np.dot(self.m1,num),np.dot(self.m2,num))
    def determinant(self):
        if np.shape(self.m1)[0]==np.shape(self.m1)[1]:
            det1=np.linalg.det(self.m1)
        else:
            print("det cannot be found")
        if np.shape(self.m2)[0]==np.shape(self.m2)[1]:
            det2=np.linalg.det(self.m2)
        else:
            print("det cannot be found")
        return (det1,det2)
           
    def matrix_mult(self):
        return np.dot(self.m1,self.m2)
    # if matrices are incompatible for multiplication value error is automatically raised
    def transpose(self):
        return ((np.transpose(self.m1)),(np.transpose(self.m2)))
    def inverse(self):
        if np.shape(self.m1)[0]==np.shape(self.m1)[1]:
            det1=np.linalg.det(self.m1)
        else:
            print("inverse cannot be found")
        if np.shape(self.m2)[0]==np.shape(self.m2)[1]:
            det2=np.linalg.det(self.m1)
        else:
            print("inverse cannot be found")
        if det1!=0:
            inv1= np.linalg.inv(self.m1)
        else:
            print("inverse cannot be found")
        if det2!=0:
            inv2= np.linalg.inv(self.m1)
        else:
            print("inverse cannot be found")
        return (inv1,inv2)
    def diagonal(self):
        return np.diagonal(self.m1)
    def rank(self):
        return (np.linalg.matrix_rank(self.m1),np.linalg.matrix_rank(self.m1))
    def eigen(self):
        if np.shape(self.m1)[0]==np.shape(self.m1)[1]:
         m1val,m1vect=np.linalg.eig(self.m1)
        else:
            print("matrix 1 is Not a square matrix")
        if np.shape(self.m2)[0]==np.shape(self.m2)[1]:
         m2val,m2vect=np.linalg.eig(self.m2)
        else:
            print("Matrix 2 is Not a square matrix ")
        print("The eigen values of M1 are ",m1val)
        print("The eigen vectors of M1 are ",m1vect)
        print("The eigen values of M2 are ",m2val)
        print("The eigen vectors of M2 are ",m2vect)
        origin = np.array([[0, 0, 0,0], [0, 0, 0,0]])
        data = np.array(list(m1vect)+list(m2vect))
        plt.quiver(*origin, data[:, 0], data[:, 1], color=['black', 'black', 'red','red'], scale=10)
        plt.show()
import random
m1 = np.array([[random.randint(1,10) for i in range(2)] for j in range(2)])
m2 = np.array([[random.randint(1,10) for i in range(2)] for j in range(2)])
matrices = Matrix(m1,m2)
print(f"Matrix 1 = {m1}, Matrix 2 = {m2}")
print("A + B = \n",matrices.elementwise_add())
print("A - B = \n",matrices.elementwise_sub())
print("A . B = \n",matrices.elementwise_mult())
print("A * k = \n",matrices.scalar_mult(6))
print("A * B = \n",matrices.matrix_mult())
print("|A| = ", round(matrices.determinant()[0],2), "|B| = ",round(matrices.determinant()[1],2))
print("A transpose = \n", matrices.transpose()[0], "\nB transpose = \n", matrices.transpose()[1])
print("A inverse = \n",matrices.inverse()[0],"\nB inverse = \n", matrices.inverse()[1])
print("Diagonal of A = ",matrices.diagonal())
print("Rank of A = ", matrices.rank()[0],"Rank of B = ",matrices.rank()[1])
matrices.eigen()

    
    

```
